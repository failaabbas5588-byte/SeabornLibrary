# Seaborn

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data =  sns.load_dataset('penguins')
data.head(5)

```

```python
plt.figure(figsize = (5, 3))
sns.lineplot(x = 'bill_length_mm', y= 'bill_depth_mm', data = data)
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.lineplot(x = 'bill_length_mm', y= 'bill_depth_mm', data = data, hue = 'sex')
plt.show()
```

```python
# hue parameter
plt.figure(figsize = (5, 3))
sns.lineplot(x = 'bill_length_mm', y= 'bill_depth_mm', data = data, hue = 'island')
plt.show()
```

```python
# size parameter
plt.figure(figsize = (5, 3))
sns.lineplot(x = 'bill_length_mm', y= 'bill_depth_mm', data = data, hue = 'sex', size = 90)
plt.show()
```

```python
# style parameter
plt.figure(figsize = (5, 3))
sns.lineplot(x = 'bill_length_mm', y= 'bill_depth_mm', data = data, hue = 'sex', style ='sex')
plt.show()
```

```python
# palrtte parameter required an argument (Accent)
# it is used for change color 
plt.figure(figsize = (5, 3))
sns.lineplot(x = 'bill_length_mm', y= 'bill_depth_mm', data = data, hue = 'sex', palette = 'Accent')
plt.show()
```

```python
# Marker

plt.figure(figsize = (5, 3))
sns.lineplot(x = 'bill_length_mm', y= 'bill_depth_mm', data = data, hue = 'sex', marker = 'o')
plt.show()
```

```python
# Dash

plt.figure(figsize = (5, 3))
sns.lineplot(x = 'bill_length_mm', y= 'bill_depth_mm', data = data, hue = 'sex')
plt.show()
```

# Histogram 

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
data =  sns.load_dataset('penguins')
data.head(5)
```

```python
plt.figure(figsize = (5, 3))
sns.distplot(data['bill_depth_mm'], color = 'r')
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.distplot(data['bill_depth_mm'],rug = True,  color = 'r')
plt.show()
```

# Scatter plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

plt.figure(figsize = (5, 3))


data =  sns.load_dataset('penguins')
data.head(5)
```

```python
plt.figure(figsize = (5, 3))
sns.scatterplot(x = 'bill_length_mm', y ='bill_depth_mm' , data = data, hue = 'sex', style = 'sex', size = 'sex', sizes = (100, 40))
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.scatterplot(x = 'bill_length_mm', y ='bill_depth_mm' , data = data, hue = 'sex', style = 'sex',palette = 'gist_ncar' )
plt.show()
```

```python
plt.figure(figsize = (5, 3))
m = {'Male':'*', 'Female' : 'o'}
sns.scatterplot(x = 'bill_length_mm', y ='bill_depth_mm' , data = data, hue = 'sex', style = 'sex',palette = 'gist_ncar' , markers = m)
plt.show()
```

# Heat Map

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
```

```python
var = np.linspace(1, 10, 20).reshape(4, 5)
var
```

```python
sns.heatmap(var, vmin = 0, vmax = 20, cmap = 'gist_heat', annot = True)
```

```python
new = np.linspace(1, 10, 10).reshape(2, 5)
sns.heatmap(new, vmax= 10, vmin = 0, cmap = 'gist_heat', annot = True)
```

```python
y = {'fontsize' : 20, 'color':'g'}
sns.heatmap(new, vmax= 10, vmin = 0, cmap = 'gist_heat', annot = True, annot_kws= y, linewidth = 8, linecolor = 'g', 
           cbar = False, xticklabels = False, yticklabels = False)
sns.set(font_scale = 5)
plt.show()
```
